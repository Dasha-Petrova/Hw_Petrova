```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with np.load('data.npz') as data:
    A,C = data ['A'], data ['C']
```

```python
def mat2vec ( A ):
    h, w = A.shape
    a = np.zeros (h *w , dtype = A.dtype)
    A = np.flipud ( A )
    for i, row in enumerate(A):
        a[i * w: i * w + w ] = row
    return a
```

```python
def vec2mat (a , shape ):
    h , w = shape
    A = np . zeros ( shape , dtype = a . dtype )
    for i in range ( h ):
        A [i, :] = a [i * w : i * w + w]
    return np.flipud(A)
```
